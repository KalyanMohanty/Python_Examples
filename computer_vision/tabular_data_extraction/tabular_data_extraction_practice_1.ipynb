```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image file 
file = r'table1.jpg'

# Read the image file using opencv
img_gray = cv2.imread(file, 0)
# Read the RGB image using opencv
img_rgb = cv2.imread(file)
```

```python
# Plot the Gray Image
plt.figure(figsize = (10,10))
plt.imshow(img_gray, cmap = 'gray')
```

```python
# plot the Original image
plt.figure(figsize = (10,10))
plt.imshow(img_rgb)
```

```python
ret, th_val = cv2.threshold(img_gray, 180, 255, cv2.THRESH_BINARY_INV)
```

```python
print(ret)
print(th_val)
plt.figure(figsize = (10,10))
plt.imshow(th_val, cmap = 'gray')
```

```python
# Dilate the image with kernal size (5,5)

ker = np.ones((5,5), np.uint8)
dil_val = cv2.dilate(th_val, ker, iterations = 1)
```

```python
print(ker)
print(dil_val)
plt.figure(figsize = (10,10))
plt.imshow(dil_val, cmap = 'gray')
```

```python
cont, hierachy = cv2.findContours(dil_val, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# print(cont)
```

```python
cordinates = []
for cnt in cont:
    x, y, w, h = cv2.boundingRect(cnt)
    cordinates.append((x, y, w, h))
    cv2.rectangle(img_rgb, (x, y), (x + w, y + h), (0, 0, 255), 1)
```

```python
plt.figure(figsize = (10, 10))
plt.imshow(img_rgb)
```

```python
cv2.namedWindow('table1_practice_1', cv2.WINDOW_NORMAL)
cv2.imwrite('table1_practice_1.jpg', img_rgb)
```
